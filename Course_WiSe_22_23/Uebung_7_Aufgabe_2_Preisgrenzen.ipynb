```python
from ipywidgets import interact, interactive, fixed, interact_manual
%matplotlib inline
import matplotlib.pyplot as plt, random
from __future__ import print_function
from google.colab import files
import ipywidgets as widgets
import pandas as pd
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = (24, 16)
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['figure.dpi'] = 32
plt.rcParams['font.size'] = 20
plt.rcParams['legend.title_fontsize'] = 20
```

```python
# Offline
"""
np.load('series_3.npy')
"""

# Online
data_df = pd.read_csv('https://github.com/trh0ly/files/blob/main/series_5.csv?raw=true',
                      index_col=0)
data = data_df.values
data.shape
```

```python
def long_call(p_T, B, c=0):
    return max(p_T - B, 0) - c

def series_fig(time):
    global lcp_range
    global series
    global strike_price

    fig = plt.figure()
    plt.plot(lcp_range, [long_call(p_T, strike_price, 0) for p_T in lcp_range], color='black', linewidth=3.5, label='Untergrenze', linestyle='--')
    plt.plot(lcp_range, lcp_range, color='black', linewidth=3.5, label='Obergrenze')
    plt.title('Wertgrenzen beim Long Call (Basispreis {}):'.format(strike_price))
    plt.xlabel('Preis des Underlyings')
    plt.ylabel('Wert Long Call')
    plt.text(strike_price, -10, 'B')
    strike_price_ = strike_price * np.exp(-(len(series) - time) * (0.01 / (len(series) / 10)))
    plt.plot(lcp_range, [long_call(p_T, strike_price_, 0) for p_T in lcp_range], color='black', linestyle=':', linewidth=3.5, label='Genauere Untergrenze (B disk. {})'.format(round(strike_price_, 1)))
    plt.ylim(-20, 250)
    plt.xlim(-20,550)
    plt.plot([], [], color='red', linewidth=2.5, label='Wert der Option')
    plt.legend(loc='best')
    plt.grid()

    plt.plot(lcp_range, series[time], color='red', linewidth=2.5)
    return()
```

```python
strike_price = 370
lcp_range = range(1,501)
series = data
```

```python
interact(series_fig, time=(1,2500,10));
```
