## Import der Programmbibliotheken

```python
%matplotlib inline
import matplotlib.pyplot as plt # Darstellung von Grafiken
import numpy as np # Ausführung mathematischer Operationen
```

## Berechnung der Payoffs von Optionen

Variablenverzeichnis:
- `p_T`: Preis des Underlyings zum Zeitpunkt der Fälligkeit `T`
- `B`: Basispreis
- `c`: Prämie Call
- `p`: Prämie Put

### Call Optionen

1. Fall: Zum Zeitpunkt `T` beträgt der Preis des Underlyings `p_T=25 GE`
<br>
2. Fall: Zum Zeitpunkt `T` beträgt der Preis des Underlyings `p_T=40 GE`

```python
# Funktion zur Berechnung des Payoffs eines Long Calls
def long_call(p_T, B, c=0):
    return max(p_T - B, 0) - c

# Funktion zur Berechnung des Payoffs eines Short Calls
def short_call(p_T, B, c=0):
    return min(B - p_T, 0) + c
```

```python
# Definition der Variablen
#################################################################################
p_T_1 = 25 # Preis Underlying; 1. Fall
p_T_2 = 40 # Preis Underlying; 2. Fall
B = 30 # Basispreis
c = 0 # Prämie
#################################################################################

print('Long Call:')
print('1. Fall: Payoff Long Call: {}'.format(long_call(p_T_1, B, c)))
print('2. Fall: Payoff Long Call: {}'.format(long_call(p_T_2, B, c)))

print('\nShort Call:')
print('1. Fall: Payoff Short Call: {}'.format(short_call(p_T_1, B, c)))
print('2. Fall: Payoff Short Call: {}'.format(short_call(p_T_2, B, c)))
```

### Put Optionen

1. Fall: Zum Zeitpunkt `T` beträgt der Preis des Underlyings `p_T=25 GE`
<br>
2. Fall: Zum Zeitpunkt `T` beträgt der Preis des Underlyings `p_T=40 GE`

```python
# Funktion zur Berechnung des Payoffs eines Long Puts
def long_put(p_T, B, p=0):
    return max(B - p_T, 0) - p

# Funktion zur Berechnung des Payoffs eines Short Puts
def short_put(p_T, B, p=0):
    return min(p_T - B, 0) + p
```

```python
# Definition der Variablen
#################################################################################
p_T_1 = 25 # Preis Underlying; 1. Fall
p_T_2 = 40 # Preis Underlying; 2. Fall
B = 30 # Basispreis
p = 0 # Prämie
#################################################################################

print('Long Put:')
print('1. Fall: Payoff Short Put: {}'.format(long_put(p_T_1, B, p)))
print('2. Fall: Payoff Short Put: {}'.format(long_put(p_T_2, B, p)))

print('\nShort Put:')
print('1. Fall: Payoff Short Put: {}'.format(short_put(p_T_1, B, p)))
print('2. Fall: Payoff Short Put: {}'.format(short_put(p_T_2, B, p)))
```

## Grafische Darstellung
### Call Optionen

```python
# Definition der Variablen
p_T_list = np.arange(0, 50, 1) # Liste mit möglichen Preisen eines Underlyings
#################################################################################
B = 30 # Basispreis
p = 2.5 # Prämie
#################################################################################

# Grafik
fig, axs = plt.subplots(1, 2)

# Long Call
axs[0].set_ylim(-15, 20)
axs[0].axhline(0, color='black')
axs[0].axvline(0, color='black')
axs[0].axvline(x=B, ymin=0.40, ymax=0.45, color='black')
axs[0].text(B-0.45, 1.75, 'B')
axs[0].plot(p_T_list, [long_call(p_T, B, p) for p_T in p_T_list]) # Payoff-Funktion
axs[0].set_title('Payoff eines Long Calls mit Basispreis {}:'.format(B))
axs[0].set_xlabel('Preis des Underlyings')
axs[0].set_ylabel('Payoff Long Call')
axs[0].grid()

# Short Call
axs[1].set_ylim(-15, 10)
axs[1].axhline(0, color='black')
axs[1].axvline(0, color='black')
axs[1].axvline(x=B, ymin=0.575, ymax=0.625, color='black')
axs[1].text(B-0.45, -1.75, 'B')
axs[1].plot(p_T_list, [short_call(p_T, B, p) for p_T in p_T_list]) # Payoff-Funktion
axs[1].set_title('Payoff eines Short Calls mit Basispreis {}:'.format(B))
axs[1].set_xlabel('Preis des Underlyings')
axs[1].set_ylabel('Payoff Short Call')
axs[1].grid()
plt.show()
```

### Put Optionen

```python
# Definition der Variablen
p_T_list = np.arange(0, 50, 1) # Liste mit möglichen Preisen eines Underlyings
#################################################################################
B = 30 # Basispreis
p = 5 # Prämie
#################################################################################

# Grafik
fig, axs = plt.subplots(1, 2)

# Long Put
axs[0].set_ylim(-10, 35)
axs[0].axhline(0, color='black')
axs[0].axvline(0, color='black')
axs[0].axvline(x=B, ymin=0.2, ymax=0.25, color='black')
axs[0].text(B-0.45, 1.75, 'B')
axs[0].plot(p_T_list, [long_put(p_T, B, p) for p_T in p_T_list]) # Payoff-Funktion
axs[0].set_title('Payoff eines Long Puts mit Basispreis {}:'.format(B))
axs[0].set_xlabel('Preis des Underlyings')
axs[0].set_ylabel('Payoff Long Put')
axs[0].grid()

# Short Put
axs[1].set_ylim(-35, 10)
axs[1].axhline(0, color='black')
axs[1].axvline(0, color='black')
axs[1].axvline(x=B, ymin=0.75, ymax=0.8, color='black')
axs[1].text(B-0.45, -3.25, 'B')
axs[1].plot(p_T_list, [short_put(p_T, B, p) for p_T in p_T_list]) # Payoff-Funktion
axs[1].set_title('Payoff eines Short Puts mit Basispreis {}:'.format(B))
axs[1].set_xlabel('Preis des Underlyings')
axs[1].set_ylabel('Payoff Short Put')
axs[1].grid()
plt.show()
```
